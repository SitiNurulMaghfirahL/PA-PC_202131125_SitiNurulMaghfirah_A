```python
import cv2 as fira
import numpy as np
import matplotlib.pyplot as plt 
```

```python
image = fira.imread ("marka_jalan.jpg")
rgb = fira.cvtColor(img, fira.COLOR_BGR2RGB)

plt.imshow(rgb)
plt.title('Gambar Asli')
```

```python
# Mengubah gambar menjadi skala abu-abu
gray = fira.cvtColor(img, fira.COLOR_BGR2GRAY)

# Menerapkan filter Gaussian untuk mengurangi noise
blur = fira.GaussianBlur(gray, (5, 5), 0)

# Menggunakan transformasi Canny untuk mendeteksi tepi
edge = fira.Canny(blur, 190, 100)

# Menerapkan transformasi Hough untuk mendeteksi garis
lines = fira.HoughLinesP(edge, 1, np.pi/150, threshold=60, minLineLength=20, maxLineGap=50)

# Menggambar garis-garis yang terdeteksi pada gambar asli
if lines is not None:
    for line in lines:
        x1, y1, x2, y2 = line[0]
        fira.line(image, (x1, y1), (x2, y2), (0, 255, 0), 3 )

citra_rgb = fira.cvtColor(image, fira.COLOR_BGR2RGB)
```

```python
#Menampilkan Garis pada Gambar 
fig, axs = plt.subplots(1,2, figsize = (10,10))
ax = axs.ravel()

ax[0].imshow(rgb)
ax[0].set_title("Gambar Asli")
ax[1].imshow(citra_rgb)
ax[1].set_title("Setelah di Deteksi")
```
